```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt   
```

```python
df = pd.read_csv('../data_export/eda_books.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Ngày xuất bản'] = pd.to_datetime(df['Ngày xuất bản'])
```

```python
df['Năm xuất bản'] = df['Ngày xuất bản'].dt.year
```

```python
df.drop(columns=['Ngày xuất bản'], inplace=True)
```

```python
sns.lineplot(data=df, x='Năm xuất bản', y='original_price', marker='o', color='blue')
plt.title('Tương quan giữa Năm xuất bản và Giá sách')
plt.xlabel('Năm xuất bản')
plt.ylabel('Giá sách')
plt.grid(True)
plt.show()
```

```python
sorted_df = df.sort_values(by='original_price', ascending=True)

sns.barplot(data = df, x='original_price', y='main_category', palette='Set1', errorbar=None, order=sorted_df['main_category'])
plt.title("Tương quan giữa Thế loại và Giá sách")
plt.xlabel("Giá sách")
plt.ylabel("Thể loại")
plt.show()
```

```python
sns.barplot(data = df, x='original_price', y='page_range', palette='Set1', errorbar=None)
plt.title("Tương quan giữa Số trang và Giá sách")
plt.xlabel("Giá sách")
plt.ylabel("Số trang")
plt.show()
```

```python
sns.barplot(data = df, x='original_price', y='Loại bìa', palette='Set1', errorbar=None)
plt.title("Tương quan giữa Loại bìa và Giá sách")
plt.xlabel("Giá sách")
plt.ylabel("Loại bìa")
plt.show()
```

```python
sns.barplot(data = df, x='original_price', y='quantity_range', palette='Set1', errorbar=None)
plt.title("Tương quan giữa Số lượt bán và Giá sách")
plt.xlabel("Giá sách")
plt.ylabel("Số trang")
plt.show()
```

```python
# Danh sách đầy đủ từ 0 đến 5 sao
order = ['0 sao', '1 sao', '2 sao', '3 sao', '4 sao', '5 sao']

# Chuyển cột rating_range thành phân loại có thứ tự
df['rating_range'] = pd.Categorical(df['rating_range'], categories=order, ordered=True)

# Vẽ biểu đồ
sns.barplot(data=df, x='original_price', y='rating_range', palette='Set1', errorbar=None)
plt.title("Tương quan giữa Số đánh giá và Giá sách")
plt.xlabel("Giá sách")
plt.ylabel("Số trang")
plt.show()
```

```python
# Danh sách đầy đủ từ 0 đến 5 sao
order = ['0 sao', '1 sao', '2 sao', '3 sao', '4 sao', '5 sao']

# Chuyển cột rating_range thành phân loại có thứ tự
df['rating_range'] = pd.Categorical(df['rating_range'], categories=order, ordered=True)

# Vẽ biểu đồ
sns.barplot(data=df, x='quantity_sold', y='rating_range', palette='Set1', errorbar=None)
plt.title("Tương quan giữa Số đánh giá và Giá sách")
plt.xlabel("Giá sách")
plt.ylabel("Số trang")
plt.show()
```

```python
numeric_cols = ['price', 'original_price', 'Số trang', 'quantity_sold', 'rating_average', 'review_count']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# 1. Tính tương quan Pearson cho các biến số
corr_vars = ['price', 'Số trang', 'quantity_sold', 'rating_average', 'review_count', 'original_price', 'Năm xuất bản']
corr_matrix = df[corr_vars].corr(method='pearson')

# In ma trận tương quan
print("Ma trận tương quan Pearson:")
print(corr_matrix['price'].sort_values(ascending=False))

# Vẽ heatmap cho ma trận tương quan
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Numeric Variables with Price')
plt.show()
```
